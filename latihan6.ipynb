```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('asli.jpeg', 0)
img2 = img.copy()
template = cv2.imread('ogr.png',0)
w, h = template.shape[::-1]

methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR',
           'cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']

plt.rcParams['figure.figsize'] = (15,15)

for met in methods:
    img = img2.copy()
    methods = eval(met)
```

```python
res = cv2.matchTemplate(img,template,methods)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
if methods in [cv2.TM_SQDIFF,cv2.TM_SQDIFF_NORMED]:
    top_left = min_loc
else:
    top_left = max_loc
bottom_right = (top_left[0] + w, top_left[1] + h)
cv2.rectangle(img, top_left, bottom_right, 255, 2)
print("hasil metode", met, ": " )
plt.subplot(121),plt.imshow(res,cmap = 'gray')
plt.title('Hasil matching'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(img,cmap = 'gray')
plt.title('Lokasi terdeteksi'), plt.xticks([]), plt.yticks([])
plt.show()
```
